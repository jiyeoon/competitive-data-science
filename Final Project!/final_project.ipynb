```python
train['date'] = pd.to_datetime(train['date'], format = '%d.%m.%Y')
```

```python
train.head(10)
```

```python
newtrain = pd.read_csv('sales_train.csv')
newtrain = newtrain[['date', 'shop_id', 'item_id', 'item_cnt_day']]
newtrain.head(10)
```

```python
year = newtrain['date'].str.split('.').str[2]
month = newtrain['date'].str.split('.').str[1]

newtrain['year'] = year
newtrain['month'] = month

newtrain.head(10)
```

```python
newtrain.groupby(['year', 'month', 'shop_id', 'item_id']).sum()['item_cnt_day']
```

```python
a = newtrain.groupby(['year', 'month', 'shop_id', 'item_id']).sum()
a.reset_index()
```

```python
newtrain1 = a.reset_index()
```

```python
!pip install sklearn
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC

import numpy as np
```

```python
newtrain1
```

```python
target = newtrain1['item_cnt_day']
newtrain2 = newtrain1[['shop_id', 'item_id']]
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

k_fold = KFold(n_splits=10, shuffle=True, random_state = 0)
```

```python
#KNN
clf = KNeighborsClassifier(n_neighbors=13)
scoring = 'accuracy'
score = cross_val_score(clf, newtrain2, target, cv=k_fold, n_jobs=1, scoring = scoring)
print(score)
```

```python
#KNN Score
round(np.mean(score) * 100, 2)
```

```python
#Decision Tree
clf = DecisionTreeClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, newtrain2, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
#Decision Tree Score
round(np.mean(score) * 100, 2)
```

```python
# Random Forest
clf = RandomForestClassifier()
scoring = 'accuracy'
score = cross_val_score(clf, newtrain2, target, cv=k_fold, n_jobs=1, scoring=scoring)
print(score)
```

```python
# Random Forest Score
round(np.mean(score) * 100, 2)
```
